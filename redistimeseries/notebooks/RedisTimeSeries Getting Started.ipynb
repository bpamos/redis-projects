```python
date = datetime.datetime.now()
currenttimestamp = int(date.timestamp() * 1000)

currenttimestamp
```

```python
myrange = 100000000
numobj = 50


for j in range(myrange):

    
    for i in range(numobj):
        obj_key = "well-"+str(i+1)
        timestamp = currenttimestamp + j
        oil = random.randint(0, 250)
        water = random.randint(0, 250)
        location = "site-"+str(random.randint(1, 5))
        # Create pipeline
        pipe = redis_obj.pipeline()
        pipe.execute_command(
            "ts.add", "ts:"+obj_key+":oil", timestamp, oil, "labels", "location", location
        )
        
        pipe.execute_command(
            "ts.add", "ts:"+obj_key+":water", timestamp, water, "labels", "location", location
        )
        
        
        # Execute pipeline
        pipe.execute()
```
